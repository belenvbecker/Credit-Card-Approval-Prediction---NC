**VIF**

```python
# eliminamos la columna 'status' de col
col = df.drop(['STATUS'], axis=1) #asumimos que col son nuestras variables independientes con las que vamos a entrenar el modelo

# Calcular el factor de inflaciÃ³n de la varianza

# Crear un dataframe para almacenar la VIF
factor  = pd.DataFrame(columns=["VIF", "Variables"] )
# por cada columna, calcular el VIF
factor["Variables"] = col.columns
factor["VIF"] = [variance_inflation_factor(col.values, i) for i in range(col.shape[1])]

# Mostrar el resultado
factor
```

La multicolinealidad ocurre cuando dos o mÃ¡s variables predictivas en el modelo estÃ¡n altamente correlacionadas.

El factor de inflaciÃ³n de varianza (VIF) es 1,0 para todas las columnas excepto niÃ±os, estado_familiar y mÃ³vil, que estÃ¡n altamente correlacionadas entre sÃ­.

AdemÃ¡s, las columnas Cantidad_hijos,Celular_propio,Cantidad_miembros_familia,Fecha_nacimiento y Anios estÃ¡n altamente correlacionadas entre sÃ­.
Por lo tanto, debemos eliminar las columnas Children, Family_status y Mobile de nuestra lista de variables independientes para obtener los mejores resultados de Random Forest Classifier.

```python
X = df.drop(['STATUS', 'Cantidad_hijos' , 'Cantidad_miembros_familia', 'Celular_propio','Fecha_nacimiento','ID'], axis = 1) # usamos a X para entrenar nuestro modelo
y = df['STATUS']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

*RandomForest*

```python
model = RandomForestClassifier(random_state=30)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
predictions =model.predict(X_test)
print("error absoluto medio: " + str(mean_absolute_error(predictions, y_test)))
```
